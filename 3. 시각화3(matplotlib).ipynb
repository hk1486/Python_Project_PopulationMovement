```python
import pandas
import matplotlib.pyplot as plt
from numpy import nan 
from matplotlib import font_manager, rc
# 그래프에 한글 오류뜰때 해결방법
font_path = './malgun.ttf' # 한글폰트 경로지정
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
```

```python
# df = pandas.read_excel('시도별 전출입 인구수.xls',header=(0,1)) # 헤더에 괄호로 묶고 인덱스 넣어주면 다중 헤더로 만들수 있음

df1 = pandas.read_excel('시도별 전출입 인구수.xls', header=0, index_col=(0,1)) # index_col=(0,1) 0,1 번 컬럼을 다중 인덱스로 만들기 

df1.loc[('전국','서울특별시')] # 다중인덱스를 조회하는 법 (튜플형태로 조회!)

df1
```

```python
df1 = df1.drop(('전출지별','전입지별'),axis=0) # 필요없는 행을 지우기
df_seoul = df1.loc['서울특별시'].drop('서울특별시',axis=0)
```

```python
sr1 = df_seoul.loc['경기도']
plt.plot(df_seoul.loc['경기도'])
```

```python
# ggplot 스타일로 그래프 만들어보기
plt.style.use('ggplot')

# 사이즈 조절하기
plt.figure(figsize=(14,5))

# x축 눈금 라벨 회전하기(90도), 사이즈 조절하기
plt.xticks(size=10,rotation=90)

# y축 눈금 라벨 회전하기(0도), 사이즈 조절하기
plt.yticks(size=10,rotation=0)

# 마커 표시 추가
plt.plot(sr1,marker='o',markersize=8)

# 타이틀 추가
plt.title('서울 -> 경기도')
# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('인구수')
# 범례 추가
plt.legend(labels = ['서울 -> 경기도'], loc = 'best')

# y축 범위 변경 (최소값,최대값)
plt.ylim(50000,800000)

# 주석 표시 (위치, 회전, 상하좌우 정렬, 사이즈)
plt.annotate('인구 이동 증가(1970-1995)',xy = (10,400000), rotation = 23, va='baseline', ha='center', fontsize=15)
# 화살표 표시 (시작점, 끝점)
plt.annotate('', xytext = (2,290000), xy = (20,620000), 
             arrowprops = dict(arrowstyle='->',color='skyblue',lw=5))
# 주석 표시 (위치, 회전, 상하좌우 정렬, 사이즈)
plt.annotate('인구 이동 감소(1995-2017)',xy = (38,540000), rotation = -11, va='baseline', ha='center', fontsize=15)
# 화살표 표시 (시작점, 끝점)
plt.annotate('', xytext = (29,620000), xy = (47,450000), 
             arrowprops = dict(arrowstyle='->',color='olive',lw=5))
```
