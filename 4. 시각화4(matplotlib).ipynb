```python
import pandas
from matplotlib import pyplot, font_manager, rc
from numpy import nan
import seaborn
```

## 한글 폰트 처리하기

```python
font_path = 'malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)
```

## 데이터 불러오고 결측치 처리하기

```python
df = pandas.read_excel('시도별 전출입 인구수.xls')
df = df.ffill() # nan값을 앞 데이터로 채움 (ex: 전출지별 - 전국)
df.head(5)
```

## '-' 데이터를 nan으로 바꾸기

```python
df.replace('-',nan,inplace=True)
df.head(5)
```

## 컬럼명 바꾸고 다중 인덱스 사용하기

```python
df.rename({'전출지별':'전출지','전입지별':'전입지'}, axis = 1, inplace = True)
df.set_index(['전출지','전입지'],inplace=True)
df.head(5)
```

## 필요없는 행 지우기 (이동자수)

```python
df.drop(('전출지별','전입지별'),axis=0,inplace=True)
df.head()
```

## 전출지가 서울인 데이터들만 뽑아내서 선 그래프 그리기

```python
df_seoul = df.loc['서울특별시']
```

## 화면 분할하여 그래프 여러개 그리기. subplot() 사용

```python
sr1 = df_seoul.loc['경기도'] # 경기도만 뽑아서 시리즈 지정

pyplot.style.use('ggplot') # 스타일 지정

figure = pyplot.figure(figsize=(15,10)) # 그래프 객체 생성(figure에 2개의 서브 플롯 생성)

ax1 = figure.add_subplot(2,1,1) # 인자 : 행크기,열크기,서브플롯순서
ax2 = figure.add_subplot(2,1,2)

ax1.plot(sr1, marker = 'o', markerfacecolor = 'orange', color='blue', markersize = 10, linewidth = 4) # 마커, 마커컬러, 선컬러, 마커사이즈, 선사이즈
ax2.plot(sr1, 'o', markersize = 8, markerfacecolor = 'skyblue',label=['서울->경기'])

ax1.legend(['서울->경기'],loc='best') # 범례표시(리스트로 작성하기), 위치지정 
ax2.legend(loc='lower right')

ax1.set_ylim(50000,800000) # y축 크기 지정
ax2.set_ylim(50000,800000)

ax1.set_title('그림1') # 그래프에 제목 달기
ax2.set_title('그림2')

ax1.set_xticklabels(sr1.index, rotation=75) # x축 눈금 라벨지정 및 75도 회전 (라벨지정필수!)
ax2.set_xticklabels(sr1.index, rotation=75) 

pyplot.show()
```

## subplot 객체 생성안할 때

```python
sr1 = df_seoul.loc['경기도']
figure = pyplot.figure(figsize=(10,5))
pyplot.plot(sr1,marker='o',markersize=12)
pyplot.title('그림3')
pyplot.ylim(50000.800000)
pyplot.legend(['서울->경기'],loc='best')
pyplot.xticks(rotation=75)
pyplot.show()
```

## 같은 화면에 그래프 추가 (서울 -> 충청남도, 경상북도, 강원도 그래프)

```python
col_years = list(map(str, range(1970,2018))) # 연도 컬럼이 문자형이기 때문에 
df3 = df_seoul.loc[['충청남도','경상북도','강원도'],col_years]

figure = pyplot.figure(figsize=(20,5))
ax = figure.add_subplot(1,1,1)
ax.plot(df3.loc['충청남도'],label = '서울->충청남도',color='olive',linewidth=2,marker='o')
ax.plot(df3.loc['경상북도'],label = '서울->경상북도',color='blue',linewidth=2,marker='o')
ax.plot(df3.loc['강원도'],label = '서울->강원도',color='red',linewidth=2,marker='o')
ax.legend(loc='best',fontsize=12)

ax.set_title('서울 -> 충청남도,경상북도,강원도',size=20)
ax.set_xlabel('기간(연도)',size=12)
ax.set_ylabel('이동 인구수',size=12)

ax.set_xticks(df3.columns) # 경고메시지방지, x축 지정
ax.set_xticklabels(df3.columns,rotation=75)

ax.tick_params(axis='x',labelsize=15) # 축 눈금 크기 지정
ax.tick_params(axis='y',labelsize=15)

pyplot.show()
```

## 한 화면 4분할 그래프

```python
col_years = list(map(str, range(1970,2018))) # 연도 컬럼이 문자형이기 때문에 
df4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'],col_years]

figure = pyplot.figure(figsize=(20,10))
ax1 = figure.add_subplot(2,2,1)
ax2 = figure.add_subplot(2,2,2)
ax3 = figure.add_subplot(2,2,3)
ax4 = figure.add_subplot(2,2,4)

ax1.plot(df4.loc['충청남도'],label = '서울->충청남도',color='olive',linewidth=2,marker='o')
ax2.plot(df4.loc['경상북도'],label = '서울->경상북도',color='blue',linewidth=2,marker='o')
ax3.plot(df4.loc['강원도'],label = '서울->강원도',color='red',linewidth=2,marker='o')
ax4.plot(df4.loc['전라남도'],label = '서울->전라남도',color='yellow',linewidth=2,marker='o')

ax1.legend(loc='best',fontsize=10)
ax1.set_title('서울->충청남도')
ax2.legend(loc='best',fontsize=10)
ax2.set_title('서울->경상북도')
ax3.legend(loc='best',fontsize=10)
ax3.set_title('서울->강원도')
ax4.legend(loc='best',fontsize=10)
ax4.set_title('서울->전라남도')
```

## 면적 그래프 그려보기, matplotlib에선 오류, pandas기본plot 을 이용

```python
df4.T.plot(kind='area',alpha=0.2,stacked=False,figsize=(20,10))

pyplot.title('서울 -> 타시도 인구이동',size=20)
pyplot.xlabel('기간',size=20)
pyplot.ylabel('이동 인구수',size=20)
pyplot.legend(loc='best',fontsize=15)
```

## 세로형 막대그래프 그려보기 (stacked=True로 한막대에 4개필드 표현가능)
- 가로형 막대그래프는 kind : barh

```python
columns_year = list(map(str,range(2000,2018)))

df4[columns_year].T.plot(kind='bar',figsize=(20,10),width=0.7,stacked=True,color=['orange','green','blue','skyblue'])

pyplot.title('서울 -> 타시도 인구이동',size=20)
pyplot.xlabel('기간',size=20)
pyplot.ylabel('이동 인구수',size=20)
pyplot.legend(loc='best',fontsize=15)
```
